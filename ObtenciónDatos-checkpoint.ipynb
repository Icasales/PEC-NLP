```python
# Archivo en blanco para guardar datos
data1 = open("/Users/icasa/Desktop/canciones1.txt", "w")
```

```python
def toma_letras(artists, c):
    x = 0
    for name in artists:
        try:
            songs = (genius.search_artist(name, max_songs=c, sort='popularity')).songs
            s = [song.lyrics for song in songs]
            data1.write("\n  \n".join(s))
            x += 1
            print(f"Canciones tomadas:{len(s)}")
        except:
            print(f"Salto en: {name}: {x}")   
```

```python
rock_artists = sample_artists + other_artists

toma_letras(rock_artists, 25)
```
