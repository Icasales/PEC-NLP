```python
import pandas as pd
dataset = pd.read_csv('/tmp/lyrics_lines.csv', delimiter= ',', index_col=False,header=None, encoding='utf-8')
```

```python
corpus = dataset[0].tolist()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tokenizo el corpus
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)

total_words = len(tokenizer.word_index) + 1

print(tokenizer.word_index)
print('El corpus tiene un total de ', total_words, ' palabras')
```

```python
#Secuencias de texto
import numpy as np
sequences = [] #Lista vacia para añadir secuencias de texto
for line in corpus:
	token_list = tokenizer.texts_to_sequences([line])[0] # inicio en 0 para iterar linea x linea, listas de lista
	for i in range(1, len(token_list)):
		n_gram_sequence = token_list[:i+1]
		sequences.append(n_gram_sequence)
```

```python
sequences[0:15]
```

```python
# Padding secuencias para igualar el largo de secuencias
# Padding 'pre' para añadir ceros al inicio
max_sequence_len = max([len(seq) for seq in sequences])
sequences = np.array(pad_sequences(sequences, maxlen=max_sequence_len, padding='pre'))
```

```python
# Secuencias de 23 digitos = max.N_palabras
sequences[0]
```

```python
# Asigno features y labels, siendo label el último dígito de cada secuencia y features los demás
features, labels = sequences[:,:-1], sequences[:,-1]

# ys categorica y one hot encoded
ys = tf.keras.utils.to_categorical(labels, num_classes=total_words)
```

```python
# variables finales para el modelo

dimensiones = 240
lstm = 200
Dropout = 0.1
```

Objetivo: accuracy >= 80%

El modelo de abajo ha sido el que ha tenido mayor exactitud, anteriormente intenté con activation = relu, añadiendo otro Dense layer, añadiendo LSTM con diferentes parametros con resultados por debajo de mi objetivo.

```python
# Construyendo mi modelo
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional, Dropout


model = Sequential()
model.add(Embedding(total_words, dimensiones, input_length=max_sequence_len-1))
model.add(Bidirectional(LSTM(lstm)))
model.add(Dropout(0.1)) # Evitar overfiting
model.add(Dense(total_words, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
Model: "sequential"
```

```python
#Fit model
#Final accuracy -> 76%

history = model.fit(features, ys, epochs=80, verbose=1)
```

```python
import time
t = time.time()

export_path_keras = "./{}.h5".format(int(t))
print(export_path_keras)

model.save(export_path_keras)
```

```python
# Función para generar canción a partir de una semilla de texto
# Para simular estructura de canción lo divido por lineas  
x = 9 # Palabras por linea
def Inicia_cancion(seed_text, next_words):
  for i in range(next_words):
    text = []
    for _ in range(x):
      token_list = tokenizer.texts_to_sequences([seed_text])[0]
      token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
      predicted_probs = model.predict(token_list)[0]
      predicted = np.random.choice([x for x in range(len(predicted_probs))], # Para evitar que repita demasiadas veces la misma palabra
                                   p=predicted_probs)
      output_word = ""
      for word, index in tokenizer.word_index.items():
        if index == predicted:
          output_word = word
          break
      
      seed_text = seed_text + ' ' + output_word
      text.append(output_word)

    seed_text = text[-1]
    text = ' '.join(text)
    print(text)
```
